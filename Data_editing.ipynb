```python
import pandas as pd
```

```python
whales = pd.read_csv('Datasets/whalescalCOFI91-06_v2.csv')
whales = whales[(whales['date_time'] < "2007-01-01 00:00:00") & (whales['date_time'] > "1996-01-01 00:00:00")]
whales = whales.sort_values(by = 'date_time', ascending = True)
whales['Date'] = pd.to_datetime(whales['date_time'], infer_datetime_format=True).dt.date
whales.head()
```

```python
krill = pd.read_csv("Datasets/krill_data2.csv")
krill['Date'] = pd.to_datetime(krill['Date'], infer_datetime_format=True)
krill.head(2)
```

```python
cast = pd.read_csv('Datasets/CalCOFI_Cast__Data.csv')
cast = cast[['Cst_Cnt', 'Date', 'Lat_Dec', 'Lon_Dec']]
cast['Date'] = pd.to_datetime(cast['Date'], infer_datetime_format=True)
cast = cast[(cast['Date'] < "2007-01-01") & (cast['Date'] > "1996-01-01")]
cast.to_csv('Datasets/Cast_96_06.csv')
cast.head()
```

```python
print(cast['Cst_Cnt'].max())
print(cast['Cst_Cnt'].min())
```

```python
bottle1 = pd.read_csv('Datasets/CalCOFI_Bottle_Data.csv')
bottle1.head()
```

```python
bottle2 = pd.read_csv('Datasets/Bottle_Subset.csv')
bottle2.head()
```

```python
bottles = pd.merge(bottle1, bottle2, on = "Btl_Cnt", how = "left")
bottles.head(2)
```

```python
cst_bottle = pd.merge(cast, bottles, on = 'Cst_Cnt')
cst_bottle.head()
```

```python
n_rows = len(cst_bottle)
for col in cst_bottle.columns:
    print('Percent Null ' + str(col) + ' : '+ str(sum(cst_bottle[col].isnull()) / n_rows))
```

Because NH3 and Light columns were mostly null they were excluded. ChlorA was imputed with the average value. The remaining rows with NaN values negligible and were dropped.

```python
cst_bottle= cst_bottle.drop(['NH3uM', 'LightP'], axis=1)
```

```python
cst_bottle['ChlorA'] = cst_bottle['ChlorA'].fillna(value = bottles['ChlorA'].mean())
```

```python
cst_bottle.dropna().to_csv('Datasets/cst_btl_FinalDataset.csv', index = False)
```
